```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
import numpy as np

import csv


def plt_hist(data, title = "Untitled", x_label = "Percent" , y_label = "Num"):
    plt.hist(data, bins = 15)
    plt.xlabel(x_label)
    plt.ylabel(y_label)

    plt.title(title)

    plt.show()

datas = pd.read_csv("prefer_result.csv")
print(datas)

plt_hist(datas["1"], "Quote "" preference")
plt_hist(datas["2"], "InequalitySign Less preference")
plt_hist(datas["3"], "Loop 'for' preference")
plt_hist(datas["5"], "TrinomialOP preference" , "times")
plt_hist(datas["6"], "C_codingstyle () preference")
```
