```python
import tokenize


#  (<,>,<=,>=)
# parameter : tokenize 
# return: Array(3) => 0: ""  , 1: ''  , 2:  

def InequalitySign_Preference(tokens):
    Less=["<","<="]
    Greater=[">",">="] 
    result=[0,0,0]
    for tok in tokens:
        if(tokenize.tok_name[tok.type] == "OP"):
            if (tok.string in Less):
                result[0] += 1
            elif(tok.string in Greater):
                result[1] += 1
    result[2]=result[0]+result[1]
    
    return result    
    

with open('hello.py', 'rb') as f:
    tokens = tokenize.tokenize(f.readline)
    \rint(InequalitySign_Preference(tokens)) 
```

```python
import tokenize
import keyword

def find_variance(tokens):
    classes=[]
    functions=[]
    variances=[]
    all=[]
    switch_c=0
    switch_f=0
    for tok in tokens:
        if(tokenize.tok_name[tok.exact_type] == "NAME"):
            if(tok.string == 'class'):
                switch_c=1
            elif(switch_c==1):
                if(tok.string not in all):
                    all.append(tok.string)
                    classes.append(tok.string)
                    switch_c=0
            if(tok.string == 'def'):
                switch_f=1
            elif(switch_f == 1):
                if(tok.string not in all):
                    all.append(tok.string)
                    functions.append(tok.string)
                    switch_f=0
            if(keyword.iskeyword(tok.string)==True):
                pass
            else:
                if(tok.string not in all):
                    all.append(tok.string)
                    variances.append(tok.string)
    return (classes, functions, variances, all)

with open('hello.py', 'rb') as f:
    tokens = tokenize.tokenize(f.readline)
    print(find_variance(tokens))
```
