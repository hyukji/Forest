```python

pd.set_option('display.max_row', 500)
pd.set_option('display.max_columns', 100)

ip = '10.180.2.80'
# ip = '114.71.103.80'
users_link = f'http://{ip}/new_db.php?q=select+*+from+users%0D%0A%0D%0A'
def code_link(user):
    before = f'http://{ip}/new_db.php?q=select+*+from+code_submission+c%2C+%28select+id%2C+max%28regtime%29+a+from+code_submission+where+id+%3D'
    after = '+group+by+p_no%29+d+where+c.id+%3D+d.id+and+d.a+%3D+c.regtime+'
    return before + user + after


 


def make_feature_df(feature_func, users, total_num_p = 78):
    df = pd.DataFrame(columns=users,index=range(1,total_num_p + 1))
    start = time()
    for user in users:
        num_p = getText(code_link(user),  '2')
        num_p = list(map(int, num_p))
        problems = getText(code_link(user),  '4')
        results=[]
        for problem in problems:
            stream = io.StringIO(problem)
            tokens = tokenize.generate_tokens(stream.readline)
            try:
                results.append(feature_func(tokens))
            except:
                results.append(None) # None = error in code

        result_i = 0
        for i, v in enumerate(num_p):
            df.loc[v][user] = results[i]
        # NaN = problem isn't submitted
    end = time()
    print(f'Running time {end - start} sec')
    return df

def avg_cleaning(df): #suji
    df.loc[:,'mean'] = df.mean(axis=1)
    try:
        df = df.divide(df['mean'], axis = 0)
    except:
        zero_index = [i for i, m in enumerate(df['mean']) if m == 0]
        for i in zero_index:
            df['mean'][i + 1] = 1
        df = df.divide(df['mean'], axis = 0)
        for i in zero_index:
            df.loc[i + 1, :] = np.nan
            
    df = df.drop(['mean'], axis=1)
    return df

def self_cleaning(df): # yeram
    num_p, num_u = df.shape
    for i in range(1, num_p + 1):
        for j in range(num_u):
            lst=df.loc[i][j]
            if type(lst) != list:
                pass
            else:
                last_num=len(lst)-1
                for k in range(len(lst)):
                    if lst[last_num] != 0:
                        lst[k]=lst[k]/lst[last_num]
                del lst[last_num]
                df.loc[i][j]=lst
    return df


# def integrate(dfs, prob, user): #seulin
#     def check_shape(dfs): # check the shape of all dfs
#         shape = dfs[0].shape
#         for df in dfs:
#             assert(shape == df.shape)
#     check_shape(dfs)
#     res = []
#     for df in dfs:
#         feature = df.loc[prob][user]
#         if type(feature) == list:
#             res.extend(feature)
#         elif type(feature) in [int, float]:
#             res.append(feature)
#         else:
#             # NaN
#             pass
#     return res



# 예람
def integrate(dfs_list):
    result_list=[]
    student_list=dfs_list[0].columns.tolist()
    index=[]
    for i in range(1, 79):
        index.append(i)
    for student in student_list:
        new_dfs=pd.DataFrame(columns=[], index=index)
        for i in range(len(dfs_list)):
            feature_name="f"+str(i)
            new_dfs[feature_name]=dfs_list[i][student]
        result_list.append(new_dfs)
        
    return result_list




# get_distribution을 위한 함수, df에서 일부 레벨 별로 df를 추출함
def get_problems_users(df, level, users):
    return df.loc[(level-1)*10+1:level*10, users]

def get_std_users(df, level, users):
    stds = []
    for u in users:
        stds.append(get_std(level, u))
    return pd.Series(stds, index=users)

# per user
def get_std(df, level): # df = prob * features
    def get_problems(df, level, user): #마지막 문제 업데이트 안 해야 함.
        #level 1 : 1~10번 문제, 2: 11~20번 문제
        return df.loc[(level-1)*10+1:level*10]
    series = get_problems(df, level)
    if series.dtype in [int, float]:
        return series.std()
    elif series.dtype == object:
        lst = series.tolist()
        new = []
        for prob in lst:
            if type(prob) == list:
                new.append(prob[0]) #나머지 항들도 같은 std 값임.
        new = pd.Series(new)
        return 1 - new.std()
#         res = []
#         for i in range(len(lst[0])): # length of each list
#             new = pd.Series([prob[i] for prob in lst])
#             res.append(new.std())
#         return res
# 유지도 : 개인별로 10개의 문제 기록을 보고 표준편차를 계산, 표준편차가 작으면 높은 가중치

def calculate_plagiarism(origin, new, weights):
    assert origin.size == new.size == weights.size
    diff = origin - new
    diff = diff.abs()
    return diff*weights
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

def distribution_df(dfs, level):
    
    def get_distribution(data, n_clustering = 3):

        mean_df = data.mean()
        np_mean_Df = mean_df.to_numpy().reshape(-1, 1)
    #     print(np_mean_Df)

        model = KMeans(n_clusters=n_clustering, init='k-means++', n_init=10, max_iter=300, tol=1e-4)
        model.fit(np_mean_Df)

        y_predict = model.labels_
        n_y_predict = len(y_predict)

        kmean_result = np.zeros(n_clustering) 

        for i in range(n_y_predict):
            kmean_result[y_predict[i]]+=1

        kmean_result /= n_y_predict

        result = [kmean_result[pred_y] for pred_y in y_predict]


        return pd.Series(result, index=list(data))


    def kmean_intergrate(s):
#         return pd.concat(s, axis=1, ignore_index=True)

        n = len(s)
        before_df = pd.concat([s[0], s[1]], axis=1)
        for i in range(2,n):
            before_df = pd.concat([before_df, s[i]], axis=1, ignore_index=True)

        return before_df

    s = []
    for df in dfs:
        temp = get_problems_users(df, level, users)
        if level == 7:
            
        temp = get_problems_users(df, level, users)
        s1 = get_distribution(temp)
        s.append(s1)
    
    return kmean_intergrate(s)
    

def seperate_df(df):
    if type(df.iloc[0, 0]) == float:
        return [df]
    
    def get_element(i):
        def help(lst):
            if type(lst) == list:
                return lst[i]
        return help
    
    res = []
    for j in range(len(df.iloc[0, 0])): # lst len
        seperated_df = df.applymap(get_element(j))
        res.append(seperated_df)
    
    return res
        
    
# make_df_list()
distribution_df(level = 5)
```
